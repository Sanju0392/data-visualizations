```python
cols_with_missing = [col for col in df.columns 
                                 if df[col].isnull().any()]
df_train=df.drop(cols_with_missing,axis=1)
```

```python
df_train
```

```python
low_cardinality_cols=[cname for cname in df_train.columns  if  
                       df_train[cname].nunique()<10 and  
                       df_train[cname].dtype== "object"]
```

```python
numeric_cols = [cname for cname in df_train.columns if 
                                df_train[cname].dtype in ['int64', 'float64']]
```

```python
my_cols = low_cardinality_cols + numeric_cols
train_predictors = df_train[my_cols]
```

```python
train_predictors
```

```python
one_hot_encoded_training_predictors = pd.get_dummies(train_predictors)
```
